#AI Foundation - Daily Assignment
Logistic Regression(Classification)

1.Create Ddataset

```python
import numpy as np
import matplotlib.pyplot as plt

# -------------------------------
# Part 1: Create Dataset
# -------------------------------
np.random.seed(0)
X = np.linspace(0, 10, 50)
y = np.array([0 if i < 5 else 1 for i in X])

# Plot dataset
plt.scatter(X, y)
plt.title("Dataset")
plt.xlabel("X")
plt.ylabel("Class")
plt.show()

```

2.Linear Model Limitation

```python
# Simple linear model
y_linear = 0.2 * X - 0.5

plt.plot(X, y_linear, color='red')
plt.scatter(X, y)
plt.title("Linear Model (Not suitable for classification)")
plt.show()
```

3.Sigmoid Function

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

z = np.linspace(-10, 10, 100)
plt.plot(z, sigmoid(z))
plt.title("Sigmoid Function")
plt.show()
```

4.Logistic Model

```python
def predict(X, w, b):
    return sigmoid(w * X + b)

# Initialize parameters
w = 0
b = 0
lr = 0.1
epochs = 1000
```

5.Training (Gradient Descent)

```python
for i in range(epochs):
    y_pred = predict(X, w, b)

    # gradients
    dw = np.mean((y_pred - y) * X)
    db = np.mean(y_pred - y)

    # update
    w -= lr * dw
    b -= lr * db

# Final prediction
y_pred = predict(X, w, b)

plt.scatter(X, y)
plt.plot(X, y_pred, color='green')
plt.title("Logistic Regression Fit")
plt.show()

```

6.Decision Boundary

```python
# boundary when sigmoid = 0.5
boundary = -b / w
print("Decision Boundary:", boundary)

plt.scatter(X, y)
plt.axvline(boundary, color='red')
plt.title("Decision Boundary")
plt.show()
```

7.Predictions

```python
y_class = [1 if i >= 0.5 else 0 for i in y_pred]

accuracy = np.mean(y_class == y)
print("Accuracy:", accuracy)
```

8.Sklearn Comparison

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X.reshape(-1,1), y)

sk_pred = model.predict(X.reshape(-1,1))

print("Sklearn Accuracy:", np.mean(sk_pred == y))
```

#CONCEPT ANSWERS

1. Why linear regression is not suitable?

Linear regression gives values outside 0–1 range, so it cannot represent probabilities.

2. Role of sigmoid function?

It converts output into probability between 0 and 1.



3. What is decision boundary?

It is the point where prediction changes from 0 to 1 (usually probability = 0.5).

4. Why optimize parameters?

To minimize error and improve prediction accuracy.

5. How does Learning rate effect training?

Too high → overshooting

Too low → slow learning

Logistic regression uses sigmoid function to convert linear output into probability. The model is trained using gradient descent. Decision boundary separates two classes effectively. Sklearn results are similar, validating implementation.
